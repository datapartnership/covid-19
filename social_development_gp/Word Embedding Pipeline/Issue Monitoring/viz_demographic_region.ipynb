```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import altair as alt
```

```python
df = pd.read_csv("F:/WB/Demographic_Prediction/twitter_demographic_100_region.csv")
df = df.rename(columns={'username':'count'})
```

```python
print('Number of Users who dont have location info: {} ({}%)'.format(df[df.location.isnull()]['count'].sum(),100*df[df.location.isnull()]['count'].sum()/df['count'].sum()))
```

```python
df.groupby('gender_class').sum()[['count']]
```

```python
df.groupby('age_class').sum()[['count']]
```

```python
df_all = df.groupbydf_all = df.groupby('location').sum()[['count']].reset_index()
df_all['count'] = 100*df_all['count']/df_all['count'].sum()

alt.Chart(df_all).mark_bar().encode(
    x = alt.X('location',sort='-y'),
    y = alt.Y('count',title = '%Users')
).properties(
    width=1000,
    height=600,
    title='Percentage of Twitter Users Accross Region'
).configure_axis(
    labelFontSize=15,
    titleFontSize=20
).configure_title(
    fontSize=30
).configure_legend(
    labelFontSize=15
).configure_axisX(
    labelAngle = -90
)
```

```python
alt.Chart(df[~df['location'].isnull()]).mark_bar().encode(
    x = alt.X('location',sort='-y'),
    y = alt.Y('count',stack="normalize", axis=alt.Axis(format='%'), title = '% Users'),
    color = 'gender_class',
    order = alt.Order('gender_class',sort='ascending'),
    tooltip = ['count']
).properties(
    width=1000,
    height=600,
    title='Twitter Users Gender Distribution Accross Region'
).configure_axis(
    labelFontSize=15,
    titleFontSize=20
).configure_title(
    fontSize=30
).configure_legend(
    labelFontSize=15
).configure_axisX(
    labelAngle = -90
)
```

```python
alt.Chart(df[~df['location'].isnull()]).mark_bar().encode(
    x = alt.X('location',sort='-y'),
    y = alt.Y('count',stack="normalize", axis=alt.Axis(format='%'), title = '% Users'),
    color = 'age_class',
    order = alt.Order('age_class',sort='ascending'),
    tooltip = ['count']
).properties(
    width=1000,
    height=600,
    title='Twitter Users Age Distribution Accross Region'
).configure_axis(
    labelFontSize=15,
    titleFontSize=20
).configure_title(
    fontSize=30
).configure_legend(
    labelFontSize=15
).configure_axisX(
    labelAngle = -90
)
```
