```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import altair as alt
```

```python
df_retweet = pd.read_csv("F:/WB/Issue Monitoring/Data/tweets_percentage_top_10_retweet_users.csv")
df_reply = pd.read_csv("F:/WB/Issue Monitoring/Data/tweets_percentage_top_10_reply_users.csv")
```

```python
retweet_dict ={}
i = 0
for user_id in df_retweet['user_id'].unique():
    for topic in ['confidence in government','economic policy','employment','food access','health care','health protocol','mobility','stigma']:
        retweet_dict[i] = {'user_id':user_id,'topic':topic,'count':df_retweet[df_retweet.user_id==user_id][topic].values[0]}
        i+=1

df_retweet_flattened = pd.DataFrame.from_dict(retweet_dict,orient='index')
```

```python
reply_dict ={}
i = 0
for user_id in df_reply['user_id'].unique():
    for topic in ['confidence in government','economic policy','employment','food access','health care','health protocol','mobility','stigma']:
        reply_dict[i] = {'user_id':user_id,'topic':topic,'count':df_reply[df_reply.user_id==user_id][topic].values[0]}
        i+=1

df_reply_flattened = pd.DataFrame.from_dict(reply_dict,orient='index')
```

```python
alt.Chart(df_retweet_flattened).mark_bar().encode(
    x = alt.X('user_id',sort='-y'),
    y = alt.Y('count',stack="normalize", axis=alt.Axis(format='%'), title = '% Users'),
    color = 'topic',
    order = alt.Order('topic',sort='ascending'),
    tooltip = ['count']
).properties(
    width=1000,
    height=600,
    title='Tweets Percentage from TOP Retweeted Users'
).configure_axis(
    labelFontSize=15,
    titleFontSize=20
).configure_title(
    fontSize=30
).configure_legend(
    labelFontSize=15
).configure_axisX(
    labelAngle = -90
)
```

```python
alt.Chart(df_reply_flattened).mark_bar().encode(
    x = alt.X('user_id',sort='-y'),
    y = alt.Y('count',stack="normalize", axis=alt.Axis(format='%'), title = '% Users'),
    color = 'topic',
    order = alt.Order('topic',sort='ascending'),
    tooltip = ['count']
).properties(
    width=1000,
    height=600,
    title='Tweets Percentage from TOP Replied Users'
).configure_axis(
    labelFontSize=15,
    titleFontSize=20
).configure_title(
    fontSize=30
).configure_legend(
    labelFontSize=15
).configure_axisX(
    labelAngle = -90
)
```
